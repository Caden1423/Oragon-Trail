```python
import random

class People():
    def __init__(self, name, money):
        self.name = name
        self.money = money
    
    def set_money(self, value):
        self.money = value
        
    def get_money(self):
        return self.money

# Function to handle events
def handle_event(event, money):
    if event == "you got robbed":
        print("You lost all your money.")
        return 0
    elif event == "you lost a kid":
        print("You lost $25.")
        return max(0, money - 25)  # Ensures money doesn't go below 0
    elif event == "no food":
        print("You lost $10.")
        return max(0, money - 10)  # Ensures money doesn't go below 0
    elif event == "your ox died":
        print("You lost $50.")
        return max(0, money - 50)  # Deduct $50 for the dead ox

day = 1  # Initialize day counter

while True:
    print("You may choose: ")
    print("\n1. Travel the trail")
    print("\n2. Learn about the trail")
    print("\n3. See the Oregon top ten")
    print("\n4. Turn off sound")
    print("\n5. Choose Management Options")
    print("\n6. End")

    choice = int(input("What is your choice? "))

    if choice == 1:
        print("\n1. Be a banker from Boston")
        print("\n2. Be a carpenter from Ohio")
        print("\n3. Be a farmer from Illinois")
        break
    elif choice == 2:
        print("Try taking a journey by covered wagon across 2000 miles of plains, rivers, and mountains.\nOn the plains, will you slosh your oxen through mud and water-filled ruts\nor will you plod through dust six inches deep")

    elif choice == 3:
        print("There is no top ten")

    elif choice == 4:
        print("Your speakers are broken now")

    elif choice == 5:
        print("You don't get that")

    elif choice == 6:
        break  # Exit the loop if the user chooses to end

    else:
        print("Invalid choice. Please choose a number between 1 and 6.")

person_choice = int(input("What person do you want to be? ")) 
money = 100
    
if person_choice == 1:
    print("You chose a banker")
    print("You are a banker, you start with $100")
elif person_choice == 2:
    print("You chose a carpenter")
    print("You are a carpenter, you start with $100")
elif person_choice == 3:
    print("You chose a farmer")
    print("You are a farmer, you start with $100")

while money > 0:
    choices = ["you got robbed", "you lost a kid", "no food", "your ox died"]
    spending = random.choice(choices)
    print("Day:", day)  # Display day counter
    print(spending)
    money = handle_event(spending, money)
    print("Your remaining money:", money)
    if day == 6:
        print("Congratulations! You've survived the Oragon trail and won the game!")
        break
    day += 1  # Increment day counter

if money <= 0:
    print("You're out of money! Game Over.")
```
